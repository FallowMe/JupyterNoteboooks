```python
#nuskaitykite auto.csv, ir pridėkite prie šios lentelės tokių stulpelių : gamintojo trys pirmos raidės didžiosiomis, suskaičiuokite, kiek kiekvienas gamintojas turėjo skirtingų automobilių modelių, ADVANCED - pavaizduokite tai grafiškai. Grafiko pobūdį ir kaip tai padarysite - renkatės patys. Hint - .groupby().reset_index() gali duoti lentelę
# kuri gali labai palengvinti užduoties atlikimą
# gamitojus imate tik 5kis, kurių yra daugiausiai

auto = pd.read_csv('auto.csv', sep='|')[['gamintojas', 'modelis', 'price', 'rida', 'kuro_tipas', 'pagaminimo_data']]

# Pridedame naują stulpelį su gamintojo trims pirmoms raidėms didžiosiomis
auto['gamintojas_3'] = auto['gamintojas'].str[:3].str.upper()
auto.head()

# Suskaičiuojame, kiek kiekvienas gamintojas turėjo skirtingų automobilių modelių
gamintojuModeliai = auto.groupby('gamintojas_3')['modelis'].nunique().reset_index()
gamintojuModeliai
# Renkamės 5 gamintojus, turinčius daugiausiai automobilių modelių
top5 = gamintojuModeliai.nlargest(5, 'modelis')
top5.tail()

#Pavaizduojame skirtingų automobilių modelių kiekį pagal gamintojus grafiniu būdu
# plt.figure(figsize=(10,6))
# plt.bar(top5['gamintojas_3'], top5['modelis'], color='skyblue')
# plt.xlabel('Gamintojas')
# plt.ylabel('Skirtingų modelių kiekis')
# plt.title('Skirtingų automobilių modelių kiekis pagal gamintoją')
# plt.show()

axis = sns.barplot(data = top5, x = 'gamintojas_3', y = 'modelis')
plt.xlabel('Gamintojas')
plt.ylabel('Skirtingų modelių kiekis')
plt.show()
```

```python
#Surinkite visus kauno dienos straipsnių pavadinimus į pandas dataframe.
#pridėkite naują stulpelį, kuriame būtų žodžių kiekis kiekviename pavadinime
#pridėkite naują stulpelį, kuriame būtų pavadinime esančių simbolių kiekis
#eksportuokite tai į CSV failą
#eksportuotą CSV failą nuskaitykite su pandas
#Koks vidutinis žodžių kiekis pavadinimuose? 
#Advanced: suraskite dažniausiai pasikartojantį žodį pavadinimuose.

kaunoDiena = pd.read_csv('KD strapsniu pav.csv', sep=';')


def splt(txt):
    return len(txt.split())

kaunoDiena['simboliu_kiekis'] = kaunoDiena['Pavadinimas'].apply(len)
kaunoDiena['zodziu_kiekis'] = kaunoDiena['Pavadinimas'].apply(splt)
kaunoDiena.head()



```

```python
vidZodziuKiekisPav = kaunoDiena['zodziu_kiekis'].mean()
print(vidZodziuKiekisPav)
```

```python
# po kiek pavadinimu turejo vienoda zodziu kieki? atsakyma pateikite vizualizacija(grafikas)

ZK = kaunoDiena.groupby('zodziu_kiekis').count().reset_index()[['zodziu_kiekis', 'Pavadinimas']]
ax = sns.barplot(data=ZK, x='zodziu_kiekis', y='Pavadinimas')
ax.set_title('Dazniausiai pasikartojantis zodis pavadinimuose')
ax.set_ylabel('Pavadinimu kiekis')
```

```python
#Advanced: suraskite dažniausiai pasikartojantį žodį pavadinimuose.

VisuZodziuSarasas = []
for e in kaunoDiena['Pavadinimas'].str.split():
    VisuZodziuSarasas.extend(e)
print(VisuZodziuSarasas)

VisiZodziai = pd.DataFrame()  # naujo data frame sukurimas
VisiZodziai['zodziai'] = VisuZodziuSarasas
# TopZ = VisiZodziai.groupby('zodziai').count().reset_index()#.sort_values(by='zodziai', ascending=False)
# TopZ.head(10)

VisiZodziai['zodziai'].value_counts().sort_values(ascending=False).reset_index()
```
