```python
dus18 = pd.read_csv('DUS2018.csv')
dus18 = dus18[['lytis', 'amzius','issilavinimas', 'bdu_spalio']]
dus18.head(3)
```

```python
# boxplot()
atl_g1 = dus18.query('issilavinimas == "G1"')['bdu_spalio'].values
atl_g2 = dus18.query('issilavinimas == "G2"')['bdu_spalio'].values
atl_g3 = dus18.query('issilavinimas == "G3"')['bdu_spalio'].values
atl_g4 = dus18.query('issilavinimas == "G4"')['bdu_spalio'].values

atl = [atl_g1, atl_g2, atl_g3, atl_g4]
edus = ['Vidurinis', 'Profesinis', 'Bakalauro', 'Magistras+']


fig, axis = plt.subplots()
#plotting code:

axis.boxplot(atl, showmeans=True, showfliers=False)

axis.set_xticklabels(edus)

#axis.legend(loc='best')
#axis.ticklabel_format(style='plain')
fig.tight_layout()
plt.show()
```

```python
# II boxplot():

fig, (left, right) = plt.subplots(1, 2)   #(1, 2) 1 eilute ir 2 stulpeliai (left, right) - pavadinimai

axis = dus18.groupby('issilavinimas').boxplot(column='bdu_spalio', subplots=False, showmeans=True, showfliers=False, ax=left)  #ax=left) grafiko perkelimas i kaire

axis.grid(visible=False)
axis.set_xticklabels(edus)

plt.show()
```

```python
# sudkite vizualizacij boxplotais 2018 met atlyginim, pagal amiaus grupes. su subplots() sukurkite 2 grafikus, kairiajame parodykite moter statistik, deinijame - vyr. Prisiminkite ir sudkite ai pavadinimus

dus18 = dus18[['lytis', 'amzius','issilavinimas', 'bdu_spalio']]
dus18.head(3)

amz_gr1 = dus18.query('amzius == "14-29"')['bdu_spalio'].values
amz_gr2 = dus18.query('amzius == "30-39"')['bdu_spalio'].values
amz_gr3 = dus18.query('amzius == "40-49"')['bdu_spalio'].values
amz_gr4 = dus18.query('amzius == "50-59"')['bdu_spalio'].values
amz_gr5 = dus18.query('amzius == "60+"')['bdu_spalio'].values

amz = [amz_gr1, amz_gr2, amz_gr3, amz_gr4, amz_gr5]
edus = ['14-29', '30-39', '40-49', '50-59', '60+']

fig, (Moterys, Vyrai) = plt.subplots(1, 2, figsize=(10, 5))

ax1 = dus18[dus18['lytis'] == 'F'].boxplot(column='bdu_spalio', by='amzius', showmeans=True, showfliers=False, ax=Moterys)
ax1.set_title('Moter atlyginimai pagal amiaus grupes')
ax1.set_ylabel('Atlyginimas')
ax1.set_xticklabels(edus)
Moterys.set_ylim(0,3000)

ax2 = dus18[dus18['lytis'] == 'M'].boxplot(column='bdu_spalio', by='amzius', showmeans=True, showfliers=False, ax=Vyrai)
ax2.set_title('Vyr atlyginimai pagal amiaus grupes')
ax2.set_ylabel('Atlyginimas')
ax2.set_xticklabels(edus)
Vyrai.set_ylim(0,3000)

plt.suptitle('2018 met atlyginimai pagal amiaus grupes ir lytis')
plt.show()
```

```python
auto = pd.read_csv('auto.csv', sep='|')
auto.head()

def konvertuotakaina(eurai):
    auto2 = float(eurai.replace(' ', ''))
    return auto2

auto['Kaina'] = auto['price'].apply(konvertuotakaina)
auto.head(5)

top5gamintojai = auto['gamintojas'].value_counts().sort_values(ascending=False).head(5).index
top5gamintojai

auto = auto.query('gamintojas in @top5gamintojai')  #@ pries kintamaji tik tada, kai jo reikia ieskoti uz lenteles ribu
auto = auto[auto['gamintojas'].isin(top5gamintojai)]
plt.figure(figsize=(10,6))
sns.boxplot(x='gamintojas', y='price', data=auto)
plt.show()
```

```python
# PAULIAUS SPRENDIMAS
auto['kaina'] = auto['price'].apply(lambda x: float(x.replace(' ', ''))) # paverciam kainas
auto_atrinkta = auto.merge(auto.value_counts('gamintojas').head(), left_on = 'gamintojas', right_on = 'gamintojas', how = 'right') # sumerginam
sns.boxplot(data=auto_atrinkta, x = 'gamintojas', y = 'kaina', showfliers = False, showmeans = True) #braizom
```

```python
#DESTYTOJO SPRENDIMAS

auto = pd.read_csv('auto.csv', sep='|')[['gamintojas','price','modelis','rida']]
auto['price'] = auto['price'].str.replace(' ','').apply(float)
g = list(auto['gamintojas'].value_counts().sort_values(ascending=False).index[:5])
sns.boxplot(data=auto.query('gamintojas in @g').sort_values(by='gamintojas'), x='gamintojas', y='price', showmeans=True, showfliers=False)
plt.show()
```

```python
url = "https://api.meteo.lt/v1/stations"


response = rq.get(url)
stations = response.json()
print(stations)
```

```python
#i to responso irinkite  sra visus ams kodus ('code' raktaodis)
stotys = []
for element in stations:   
    stotys.append(element['code'])
print(stotys)
```

```python
#atrinkiti vis stoi ufiksuot temperatr 00:00:00 laiku, iai dienai (# surinkti  sra fiksuotas temperatras iandien, 00:00:00 laiku)

visosTemp = []

for stotis in stotys:
    url = f'https://api.meteo.lt/v1/stations/{stotis}/observations/2024-04-18'
    #print(url) cia matosi su print kaip veikia f stringas
    response = rq.get(url)
    temp = response.json()
    sar = temp['observations'][0]['airTemperature']  # paima [0] indeksa observation ir salia airTemperature esancia verte
    visosTemp.append(sar)  
print(visosTemp)
```

```python
#rasti 3 meteo stotis, kuriose fiksuota temperatra 00:00 laiku buvo didiausia:
df = pd.DataFrame() # sukuriamas tuscias data frame
df['stotis'] = stotys
df['temperatura'] = visosTemp
df.sort_values(by='temperatura', ascending=False).head(3)
```
